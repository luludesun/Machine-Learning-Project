## Label Encoding and Normalize Data


```python
le = LabelEncoder()
# adult = adult.apply(le.fit_transform)
# covtype = covtype.apply(le.fit_transform)
# letter = letter.apply(le.fit_transform)

def label(data):
    for i in range(0,data.shape[1]):
        if type(data[0, i]) == str:
            le.fit(data[:,i])
            data[:,i] = le.transform(data[:,i])
            
            
label(adult)
label(covtype)
label(letter)
label(rain)

def normalize(data):
    for i in range(0,data.shape[1] - 1):
        data[:,i] = (data[:,i] - np.min(data[:,i])) / (np.max(data[:,i]) - np.min(data[:,i]))

# Normalize
normalize(adult)
normalize(covtype)
normalize(letter)
normalize(rain)
```

## Shuffle and Split Dataset 

```python
data = [adult, covtype, letter, rain]
data_names = ['adult', 'covtype', 'letter', 'rain']

def split(data):
    X = data[:, 0:data.shape[1] -1]
    Y = data[:,data.shape[1]-1, np.newaxis]
    Y = Y.astype('int')
    
    x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size = 5000, shuffle = True)
    
    return x_train, x_test, y_train, y_test
```

## Hyperparameter


```python
svm = [[] * 4]
svm_hyp = []

for i in range(4):
    dataset = data[i]
    x_train, x_test, y_train, y_test = split(dataset)
    
    param = {'kernel': ['rbf', 'poly','linear'],'C': [0.01,0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001], 'degree': [2, 3, 4]}
    svm = SVC(probability = True)
#     clf = GridSearchCV(svm, param, n_jobs = -1, verbose=2, cv = StratifiedKFold(n_splits = 5), scoring=['accuracy', 'roc_auc_ovr', 'f1_micro'], refit = False)
    clf = GridSearchCV(svm, param, n_jobs = -1, verbose=2, cv = StratifiedKFold(n_splits = 5))
    clf.fit(x_train, np.ravel(y_train, order = 'C'))
    svm_hyp.append(clf.best_params_)
 
```

```python
svm_hyp
```

```python
knn = [[] * 4]
knn_hyp = []

for i in range(4):
    dataset = data[i]
    x_train, x_test, y_train, y_test = split(dataset)
    
    param = {'weights': ['distance', 'uniform'], 'algorithm': ['auto', 'ball_tree', 'kd_tree'],'n_neighbors': [1, 3, 5, 20, 35, 50, 70, 100, 200, 400, 600]}   
    knn = KNeighborsClassifier()
#     clf = GridSearchCV(knn, param, n_jobs = -1, verbose=2, cv = StratifiedKFold(n_splits = 5), scoring=['accuracy', 'roc_auc_ovr', 'f1_micro'],refit = False)
    clf = GridSearchCV(knn, param, n_jobs = -1, verbose=2, cv = StratifiedKFold(n_splits = 5))
    clf.fit(x_train, np.ravel(y_train, order = 'C'))
    knn_hyp.append(clf.best_params_)
    
```

```python
knn_hyp
```

```python
dt = [[] * 4]
dt_hyp = []

for i in range(4):
    dataset = data[i]
    x_train, x_test, y_train, y_test = split(dataset)
    
    param = {'criterion': ['gini', 'entropy'], 'splitter' : ['best', 'random'], 'max_depth' : [2, 5, 10, 50, 100, 200, 400]}
    dt = DecisionTreeClassifier()
#     clf = GridSearchCV(dt, param, n_jobs = -1, verbose=2, cv = StratifiedKFold(n_splits = 5), scoring=['accuracy', 'roc_auc_ovr', 'f1_micro'], refit = False)
    clf = GridSearchCV(dt, param, n_jobs = -1, verbose=2, cv = StratifiedKFold(n_splits = 5))
    clf.fit(x_train, np.ravel(y_train, order = 'C'))
    dt_hyp.append(clf.best_params_)
```

```python
dt_hyp
```

```python
#svm
for i in range(4):
    dataset = data[i]
    name = data_names[i]
    
    for j in range(5):
        x_train, x_test, y_train, y_test = split(dataset)
        
        svm = SVC(**svm_hyp[i])
        svm.fit(x_train, np.ravel(y_train, order = 'C'))
        perf[name]['svm'].append(svm.score(x_test,y_test))
```

```python
#knn
for i in range(4):
    dataset = data[i]
    name = data_names[i]
    
    for j in range(5):
        x_train, x_test, y_train, y_test = split(dataset)
        
        knn = KNeighborsClassifier(**knn_hyp[i])
        knn.fit(x_train, np.ravel(y_train, order = 'C'))
        perf[name]['knn'].append(knn.score(x_test,y_test))
```

```python
#dt
for i in range(4):
    dataset = data[i]
    name = data_names[i]
    
    for j in range(5):
        x_train, x_test, y_train, y_test = split(dataset)
        
        dt = DecisionTreeClassifier(**dt_hyp[i])
        dt.fit(x_train, np.ravel(y_train, order = 'C'))
        perf[name]['dt'].append(dt.score(x_test,y_test))
```

## Comparison 

```python
#Sorting t test by algorithm
for i in range(3):
    for j in range(i+1, 3):
        a = perf_algorm[algorm_name[i]]
        b = perf_algorm[algorm_name[j]]
        print('T test between %s and %s' %(algorm_name[i], algorm_name[j]))
        print(sp.stats.ttest_ind(a,b))
```
